# Deutsch-Jozsa Algorithm

## Problem Statement:
Input: <br>
        Boolean function (f) that is balanced (f(0) != f(1)) or constant (f(0) = f(1)) <br>
        Binary String of 1's and 0's <br>
        
Output:<br>
The system is supposed to determine whether the function is a balanced of constant

## Classical Solution

For an n-bit input string, a classical computer will take:<br>
    Best Case: 2 calls to f (if it is a balanced function)<br>
    Worst Case: $\frac{n}{2} + 1$ calls to the f (if it is constant)

## Quantum Solution
The Deutsch-Jozsa Algorithm can solve this problem in exactly 1 call to the Quantum Oracle and $n+1$ qubits.

```python
import numpy as np
from qiskit import IBMQ, Aer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, assemble, transpile
from qiskit.visualization import plot_histogram
import warnings
warnings.filterwarnings('ignore')
```

```python
# Size of input string
n = 3
```

```python
# Constant Oracle
'''
    if f(x) = 0: apply I-gate to qubit in register 2 
    else: apply X-gate to qubit register 2
'''
oracle_const = QuantumCircuit(n+1)
out = np.random.randint(2)
if (out == 1):
    oracle_const.x(n)
fig = oracle_const.draw('mpl')
```

```python
# Balanced Oracle
oracle_bal = QuantumCircuit(n+1)
str = '101'
for qubit in range(len(str)):
    if (str[qubit] == '1'):
        oracle_bal.x(qubit)
oracle_bal.barrier()

for qubit in range(n):
    oracle_bal.cx(qubit, n)
oracle_bal.barrier()

for qubit in range(len(str)):
    if (str[qubit] == '1'):
        oracle_bal.x(qubit)
    
fig = oracle_bal.draw('mpl')
```

```python
circuit = QuantumCircuit(n+1, n)
for qubit in range(n):
    circuit.h(qubit)
    
circuit.x(n)
circuit.h(n)
circuit += oracle_bal

for qubit in range(n):
    circuit.h(qubit)
circuit.barrier()

for i in range(n):
    circuit.measure(i, i)
fig = circuit.draw('mpl')
fig.savefig("../Images/Deutsch-Jozsa_circuit.jpg")
```

```python
# use local simulator
aer_sim = Aer.get_backend('aer_simulator')
qobj = assemble(circuit, aer_sim)
results = aer_sim.run(qobj).result()
answer = results.get_counts()

plot = plot_histogram(answer)
```
